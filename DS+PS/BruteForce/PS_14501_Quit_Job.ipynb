```python
def maxProfCalc(N, tp_arr):
    dp = [0 for i in range(N+1)] # 일할 수 있는 날짜를 담을 리스트

    for i in range(N, 0, -1): # N부터 -1씩 -1에 다다를때까지 반복
        temp = i + tp_arr[i][0]
        if i == N:
            if tp_arr[i][0] == 1:
                dp[i] = tp_arr[i][1]
        elif temp == N + 1:
            dp[i] = max(tp_arr[i][1], dp[i + 1])
        elif temp < N + 1:
            dp[i] = max(tp_arr[i][1] + dp[temp], dp[i + 1])
        else:
            dp[i] = dp[i+1]
    print(dp[1])
        
if __name__ == '__main__':
    N = int(input()) # 남은 N일을 입력받음
    tp_arr = [] # t와 p를 한꺼번에 튜플로 받기 위한 빈 리스트
    tp_arr.append(()) # 리스트를 1부터 시작하기 위해 빈 튜플 삽입
    maxProfit = 0 
    
    for i in range(N):
        ti, pi = map(int, input().split())
        tp_arr.append((ti, pi))
    print(tp_arr)
        
    maxProfCalc(N, tp_arr)
```
